```python
import numpy as np
N = int(input("Enter the number of coordinates: "))
x_values = []
y_values = []
for i in range(N):
    x, y = map(int, input("Enter the coordinate (X,Y): ").split())
    x_values.append(x)
    y_values.append(y)
x_values = np.array(x_values)
y_values = np.array(y_values)
# Perform the Least Squares fit
A = np.vstack([x_values, np.ones(N)]).T
m, c = np.linalg.lstsq(A, y_values, rcond=None)[0]
print("Equation of the fitted line: y =", m, "* x +", c)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Take input for the number of coordinates
N = int(input("Enter the number of coordinates (N > 5): "))

if N <= 5:
    print("Please enter a value of N greater than 5.")
    exit()

# Initialize arrays to store the X and Y values
x_values = []
y_values = []

# Take input for each coordinate pair
for i in range(N):
    x, y = map(int, input("Enter the coordinate (X,Y): ").split())
    x_values.append(x)
    y_values.append(y)

# Convert the lists to numpy arrays
x_values = np.array(x_values)
y_values = np.array(y_values)

# Fit a curve using numpy's polynomial curve fitting
degree = N - 1  # Use polynomial degree equal to N-1 for a perfect fit
coefficients = np.polyfit(x_values, y_values, degree)

# Generate the curve using the fitted coefficients
curve_x = np.linspace(min(x_values), max(x_values), 100)
curve_y = np.polyval(coefficients, curve_x)

# Plot the original data points and the fitted curve
plt.scatter(x_values, y_values, label='Data Points')
plt.plot(curve_x, curve_y, color='r', label='Fitted Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Curve Fitting using Least Squares')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = int(input("Enter the number of coordinates (N > 5): "))
if N <= 5:
    print("Please enter a value of N greater than 5.")
    exit()
x_values = []
y_values = []
for i in range(N):
    x, y = map(int, input("Enter the coordinate (X,Y): ").split())
    x_values.append(x)
    y_values.append(y)
x_values = np.array(x_values)
y_values = np.array(y_values)
degree = N - 1
coefficients = np.polyfit(x_values, y_values, degree)
curve_x = np.linspace(min(x_values), max(x_values), 100)
curve_y = np.polyval(coefficients, curve_x)
plt.scatter(x_values, y_values, label='Data Points')
plt.plot(curve_x, curve_y, color='r', label='Fitted Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Curve Fitting using Least Squares')
plt.show()
equation = "y = "
for i in range(degree, 0, -1):
    equation += f"{coefficients[degree - i]} * x^{i} + "
equation += f"{coefficients[degree]}"
print("Equation of the fitted curve:", equation)
```
